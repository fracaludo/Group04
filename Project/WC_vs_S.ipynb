```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
file_path = '/Users/ludovica/Documents/GitHub/Group04/Project/WC_ S.csv'  # Replace with the path to your CSV file
df = pd.read_csv(file_path)

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df["Selectivity"], df["Working Capacity [mol/kg]"], color='blue', marker='x')

# Label each point with its corresponding MOF name
for i, row in df.iterrows():
    plt.annotate(row["MOF"], (row["Selectivity"], row["Working Capacity [mol/kg]"]),
                 textcoords="offset points", xytext=(0, 5), ha='center')

# Customize plot
plt.title('Selectivity vs Working Capacity')
plt.xlabel('Selectivity')
plt.ylabel('Working Capacity [mol/kg]')
plt.xlim(0, 35)  # Adjust x-limit as necessary
plt.ylim(0, 3)   # Adjust y-limit as necessary
plt.tight_layout()
plt.show()
```
