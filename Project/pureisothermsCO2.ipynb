```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV files into pandas DataFrames
data_utsa20 = pd.read_csv('/Users/emmak/github/project - utsa20-co2.csv')
data_utsa80 = pd.read_csv('/Users/emmak/github/project - co2 utsa80.csv')
data_nott300 = pd.read_csv('/Users/emmak/github/project - co2 nott300.csv')
data_mgmof = pd.read_csv('/Users/emmak/github/project - co2-mgmof.csv')
data_zif8 = pd.read_csv('/Users/emmak/github/project - co2-zif8.csv')

# Extract the relevant columns from each DataFrame
# Pressure is measured in bar
pressure_utsa20 = data_utsa20['pressure(bar)']
pressure_utsa80 = data_utsa80['pressure(bar)']
pressure_nott300 = data_nott300['pressure(bar)']
pressure_mgmof = data_mgmof['pressure(bar)']
pressure_zif8 = data_zif8['pressure(bar)']

# Loading values are the amount of gas adsorbed in mol/kg
loading_utsa20 = data_utsa20['loading_abs_avg(mol/kg)']
deviation_utsa20 = data_utsa20['loading_abs_dev(mol/kg)']  # Error in loading for utsa20

loading_utsa80 = data_utsa80['loading_abs_avg(mol/kg)']
deviation_utsa80 = data_utsa80['loading_abs_dev(mol/kg)']  # Error in loading for utsa80

loading_nott300 = data_nott300['loading_abs_avg(mol/kg)']
deviation_nott300 = data_nott300['loading_abs_dev(mol/kg)']  # Error in loading for nott300

loading_mgmof = data_mgmof['loading_abs_avg(mol/kg)']
deviation_mgmof = data_mgmof['loading_abs_dev(mol/kg)']  # Error in loading for mgmof

loading_zif8 = data_zif8['loading_abs_avg(mol/kg)']
deviation_zif8 = data_zif8['loading_abs_dev(mol/kg)']  # Error in loading for zif8

# Set global font size using rcParams
plt.rcParams.update({'font.size': 16})  # Adjust this value as needed

# Create a figure for the plot
plt.figure(figsize=(10, 8))

# Plot the CO2 utsa20 isotherm data with error bars and a line connecting the points
plt.errorbar(pressure_utsa20, loading_utsa20, yerr=deviation_utsa20, 
             fmt='o', color='orange', label='CO₂ pure isotherm and absolute deviation for UTSA-20', capsize=5)
plt.plot(pressure_utsa20, loading_utsa20, color='orange', linestyle='-')

# Plot the CO2 utsa80 isotherm data with error bars and a line connecting the points
plt.errorbar(pressure_utsa80, loading_utsa80, yerr=deviation_utsa80, 
             fmt='o', color='b', label='CO₂ pure isotherm and absolute deviation for UTSA-80', capsize=5)
plt.plot(pressure_utsa80, loading_utsa80, color='b', linestyle='-')

# Plot the CO2 nott300 isotherm data with error bars and a line connecting the points
plt.errorbar(pressure_nott300, loading_nott300, yerr=deviation_nott300, 
             fmt='o', color='r', label='CO₂ pure isotherm and absolute deviation for NOTT-300', capsize=5)
plt.plot(pressure_nott300, loading_nott300, color='r', linestyle='-')

# Plot the CO2 mgmof isotherm data with error bars and a line connecting the points
plt.errorbar(pressure_mgmof, loading_mgmof, yerr=deviation_mgmof, 
             fmt='o', color='g', label='CO₂ pure isotherm and absolute deviation for Mg-MOF-74', capsize=5)
plt.plot(pressure_mgmof, loading_mgmof, color='g', linestyle='-')

# Plot the CO2 zif8 isotherm data with error bars and a line connecting the points
plt.errorbar(pressure_zif8, loading_zif8, yerr=deviation_zif8, 
             fmt='o', color='c', label='CO₂ pure isotherm and absolute deviation for ZIF-8', capsize=5)
plt.plot(pressure_zif8, loading_zif8, color='c', linestyle='-')

# Label the x-axis (Pressure) and y-axis (Loading), making sure to include the units
plt.xlabel('Pressure [bar]', fontsize=18)  # Pressure measured in bar
plt.ylabel('Loading Absolute Average [mol/kg]', fontsize=18)  # Loading measured in mol/kg

# Add a legend to differentiate between all data
plt.legend(fontsize=14)

# Add a title to the plot, describing the content of the graph
plt.title('CO₂ Pure Isotherms for the different MOFs: Loading [mol/kg] vs Pressure [bar]', fontsize=18)

# Display and save the plot
plt.savefig('isothermeCO2.png', dpi=600)
plt.show()
```
