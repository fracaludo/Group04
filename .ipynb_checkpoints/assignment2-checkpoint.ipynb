```python
import pyiast
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


file_path_CH4 = '/Users/emmak/github/CH4_IRMOF1.csv'
file_path_CO2 = '/Users/emmak/github/CO2_IRMOF1.csv'
df_co2 = pd.read_csv(file_path_CO2)
df_ch4 = pd.read_csv(file_path_CH4)

df_co2.head(), df_ch4.head()

ch4_isotherm = pyiast.InterpolatorIsotherm(df_ch4,
                                    loading_key="Loading(mmol/g)",
                                    pressure_key="Pressure(bar)")

co2_isotherm = pyiast.InterpolatorIsotherm(df_co2,
                                    loading_key="Loading(mmol/g)",
                                    pressure_key="Pressure(bar)")

# Pressure range from 0.1 to 4.0 bar, with 0.2 bar increments
pressures = np.arange(0.1, 4.1, 0.2)

# Mole fractions for the biogas mixture: 60% CH4, 40% CO2
y_co2 = 0.4  # 40% CO2
y_ch4 = 0.6  # 60% CH4

# Initialize lists to store adsorption data
adsorption_co2 = []
adsorption_ch4 = []

# Loop through each pressure and perform IAST to calculate mixture adsorption
for P in pressures:
    # Perform IAST for the binary mixture
    adsorbed_amounts = pyiast.iast([y_co2, y_ch4], [co2_isotherm, ch4_isotherm], P)
    
    # Store the adsorption of each component
    adsorption_co2.append(adsorbed_amounts[0])  # CO2 adsorption at this pressure
    adsorption_ch4.append(adsorbed_amounts[1])  # CH4 adsorption at this pressure

# Plotting the adsorption isotherms for CO2 and CH4
plt.plot(pressures, adsorption_co2, label='CO2 binary isotherm')
plt.plot(pressures, adsorption_ch4, label='CH4 binary isotherm')
plt.xlabel('Pressure [bar]')
plt.ylabel('Loading [mmol/g]')
plt.legend()
plt.title('Binary Mixture Isotherms for CO2 and CH4 (60% CH4, 40% CO2)')
plt.show()
print(f"Pressure: {P} bar, CO2 Adsorbed: {adsorbed_amounts[0]} mmol/g, CH4 Adsorbed: {adsorbed_amounts[1]} mmol/g")
```

```python
# Pressures at which to calculate selectivity (0.1, 1, 2, and 3 bars)
selectivity_pressures = [0.1, 1, 2, 3]

# List to store selectivity values
selectivities = []

# Loop through the selectivity pressures and calculate selectivity
for P in selectivity_pressures:
    # Perform IAST at each pressure
    adsorbed_amounts = pyiast.iast([y_co2, y_ch4], [co2_isotherm, ch4_isotherm], P)
    
    q_co2 = adsorbed_amounts[0]  # Amount of CO2 adsorbed
    q_ch4 = adsorbed_amounts[1]  # Amount of CH4 adsorbed
    
    # Calculate selectivity (S_CO2/CH4)
    selectivity = (q_co2 / q_ch4) / (y_co2 / y_ch4)
    selectivities.append(selectivity)

# Display the selectivity results at different pressures
for i, P in enumerate(selectivity_pressures):
    print(f"Selectivity of CO2 over CH4 at {P} bar: {selectivities[i]:.2f}")
```

```python
# Now, let's check the selectivity when y_CO2 drops to 0.2 and y_CH4 becomes 0.8
y_co2_new = 0.2  # 20% CO2
y_ch4_new = 0.8  # 80% CH4

# List to store new selectivity values
selectivities_new = []

# Loop through the same pressures to check selectivity with new mole fractions
for P in selectivity_pressures:
    # Perform IAST with new mole fractions
    adsorbed_amounts = pyiast.iast([y_co2_new, y_ch4_new], [co2_isotherm, ch4_isotherm], P)
    
    q_co2 = adsorbed_amounts[0]  # Amount of CO2 adsorbed
    q_ch4 = adsorbed_amounts[1]  # Amount of CH4 adsorbed
    
    # Calculate selectivity with the new mole fractions
    selectivity_new = (q_co2 / q_ch4) / (y_co2_new / y_ch4_new)
    selectivities_new.append(selectivity_new)

# Display the new selectivity results
print("\nSelectivity with y_CO2 = 0.2 and y_CH4 = 0.8:")
for i, P in enumerate(selectivity_pressures):
    print(f"Selectivity of CO2 over CH4 at {P} bar: {selectivities_new[i]:.2f}")
```
