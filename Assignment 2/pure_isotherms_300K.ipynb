```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV file into a pandas DataFrame
# The file contains isotherm data for CH₄ and CO₂
data = pd.read_csv('/Users/ludovica/Documents/Bachelor/BA5/Modeling Lab/Assignment 2 /isotherm_data - Feuille 1.csv')  # Replace with the actual file path if needed

# Extract the relevant columns from the DataFrame
# Pressure is measured in bar
pressure_ch4 = data['pressure_ch4(bar)']

# Loading values are the amount of gas adsorbed in mol/kg
loading_ch4 = data['loading_absolute_average_ch4(mol/kg)']
deviation_ch4 = data['loading absolute_dev_ch4(mol/kg)']  # Error in loading for CH₄

loading_co2 = data['loading_absolute_average_co2(mol/kg)']
deviation_co2 = data['loading absolute_dev_co2(mol/kg)']  # Error in loading for CO₂

# Create a figure for the plot
plt.figure(figsize=(8, 6))

# Plot the CH₄ isotherm data with error bars representing deviation
plt.errorbar(pressure_ch4, loading_ch4, yerr=deviation_ch4, 
             fmt='o', color='b', label='CH₄ pure isotherm and absolute deviation', capsize=5)

# Plot the CO₂ isotherm data with error bars
plt.errorbar(pressure_ch4, loading_co2, yerr=deviation_co2, 
             fmt='o', color='r', label='CO₂ pure isotherm and absolute deviation', capsize=5)

# Label the x-axis (Pressure) and y-axis (Loading), making sure to include the units
plt.xlabel('Pressure [bar]')  # Pressure measured in bar
plt.ylabel('Loading Absolute Average [mol/kg]')  # Loading measured in mol/kg

# Add a legend to differentiate between CH₄ and CO₂ data
plt.legend()

# Add a title to the plot, describing the content of the graph
plt.title('CH₄ and CO₂ Pure Isotherms: Loading [mol/kg] vs Pressure [bar]')

# Display the plot
plt.show()
```

Above is shown the plot for the amount of gas absorbed by the MOF (loading) [mol/kg] as a function of the pressure [bar] with their corresponding error bars. More information is provided on the paper.
